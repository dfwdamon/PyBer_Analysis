```python
# Import dependencies.
import matplotlib.pyplot as plt
```

### 5.1.5 Create Bar Charts Using the MATLAB Approach

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create a vertical bar chart plot
plt.bar(x_axis, y_axis)
```

```python
# create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create title
plt.title("PyBer Fare by Month")
#legend
plt.grid() #! cant mix in a ax.grid()
plt.legend()
```

```python
# 5.1.5 # create a horizontal bar chart plot 
plt.barh(x_axis, y_axis)
```

```python
plt.barh(y_axis, x_axis)
#plt.bar(y_axis, x_axis)
```

### To invert the y-axis to have the months in ascending order, use the gca() method.

```python
# Create a plot with January @ top, Dec @ bottom.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
#plt.gca().invert_xaxis()
```

```python

# create the plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")
#create title
plt.title("PyBer Fare by Month")
#legend
plt.grid()

plt.gca().invert_yaxis()
plt.legend()
```

### 5.1.6 Create charts with OO Approach

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

#### OO Interface https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html

```python
# create horizontal bar chart plot with ax.plt()
fig, ax = plt.subplots()
#ax.barh(x_axis, y_axis)
ax.barh(y_axis, x_axis)
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
#ax.barh(y_axis, x_axis)
plt.gca().invert_yaxis()
plt.legend()    
```
